```python
import os
from mlflow.tracking import MlflowClient
from dotenv import load_dotenv
import mlflow

load_dotenv()

# FIRST: set tracking URI
mlflow.set_tracking_uri(os.getenv("MLFLOW_SERVER"))

# THEN: create client
client = MlflowClient()

registry_name = os.getenv("MLFLOW_REGISTRY_NAME")
alias = os.getenv("ENV")

print(f"Registry: {registry_name}, Alias: {alias}")

try:
    version = client.get_model_version_by_alias(registry_name, alias)
    print(f"Alias '{alias}' → version {version.version}, run_id={version.run_id}")
except Exception as e:
    print(f"Error fetching alias: {e}")
```
